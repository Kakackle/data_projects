```python
genre = 'pop'

df_path = f'dataframes/with_counts/{genre}_count.csv'

df_with = pd.read_csv(df_path, index_col=0)
```

```python
male_artists = [
    'jay-z', 'kendrick lamar', 'mf doom', 'kanye west',
    'eminem', '2pac', 'ice cube', 'nas', 'a tribe called quest', 'mos def',

    'queens of the stone age', 'nirvana', 'foo fighters', 'the strokes',
    'the white stripes', 'led zeppelin', 'fugazi', 'the rolling stones',
    'black sabbath', 'deep purple',

    'michael jackson',

    'stevie wonder', 'marvin gaye', 'curtis mayfield', 'otis redding',
    'al green', 'sam cooke', 'clarence carter'
]

female_artists = [
    'taylor swift', 'dua lipa', 'carly rae jepsen', 'madonna', 'charli xcx',
    'olivia rodrigo', 'beyoncé', 'aaliyah', 'britney spears',

    'aretha franklin', 'nina simone', 'minnie riperton'
]
```

```python
df_with['gender'] = 'male'
```

```python
def add_gender(df_with):
    df_with['gender'] = 'male'
    for i in df_with.index:
        if df_with['Artist'][i].lower() in male_artists:
            df_with['gender'][i] = 'male'
        if df_with['Artist'][i].lower() in female_artists:
            df_with['gender'][i] = 'female'
    return df_with
```

```python
for genre in ['pop', 'rock', 'rap', 'soul']:
    group_stats_sum = f'dataframes/group_stats/sum/{genre}_10.csv'
    df1 = pd.read_csv(group_stats_sum, index_col = 0)
    group_stats_mean = f'dataframes/group_stats/sum/{genre}_10.csv'
    df2 = pd.read_csv(group_stats_mean, index_col = 0)
    with_counts = f'dataframes/with_counts/{genre}_count.csv'
    df3 = pd.read_csv(with_counts, index_col = 0)

    dfs = [df1, df2, df3]
    paths = [group_stats_sum, group_stats_mean, with_counts]

    for i,df in enumerate(dfs):
        df = add_gender(df)
        df.to_csv(paths[i])
```
