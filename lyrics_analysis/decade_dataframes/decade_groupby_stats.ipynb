```python
import pandas as pd
```

```python
counts_df = pd.read_csv('decades_counts.csv', index_col=0)
```

cols

```python
count_cols = ['manual_love_count',
 'manual_swears_count',
 'manual_money_count',
 'manual_violence_count',
 'manual_drugs_count',
 'manual_gendered_count',
 'manual_sadness_count',
 'manual_joy_count',
 'manual_yes_count',
 'manual_no_count']
```

```python
to_save_cols = [*count_cols, 'unique_words', 'total_words',
                'featured_count','producer_count', 'writer_count']
```

```python
df_sum_artist = counts_df.groupby(['Artist'])[to_save_cols].sum()
```

```python
df_sum_artist
```

```python
df_mean_artist = counts_df.groupby(['Artist'])[to_save_cols].mean()
```

```python
df_mean_artist.sample(3)
```

```python
df_sum_artist.to_csv('artist_groupby_sum.csv')
```

```python
df_mean_artist.to_csv('artist_groupby_mean.csv')
```
