```python
from dash import Dash, html, dcc, callback, Output, Input
import dash_bootstrap_components as dbc
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('dataframes/with_counts/soul_count.csv')
```

```python
app = Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])

artists = list(df.Artist.unique())

items = [{"label": artist, "value": artist} for artist in artists]

app.layout = html.Div([
    html.H1(children = 'First Dash App', style={'textAlign': 'center'}),
    # dcc.Dropdown(df.Artist.unique(), 'Al Green', id='artist-selection'),
    # dbc.DropdownMenu(label="Artist", id="artist-selection", children=items),
    dbc.Select(id="artist-selection", options=items),
    dcc.Graph(id='graph-content')
])
```

```python
@callback(
    Output('graph-content', 'figure'),
    Input('artist-selection', 'value')
)
def update_graph(selected_artist):
    dff = df[df.Artist == selected_artist]
    fig = px.scatter(dff, x='Year', y='sadness')
    return fig

# if __name__ = '__main__':
#     app.run(debug=True)
app.run()
```
