```python
import wordcloud
from wordcloud import WordCloud, STOPWORDS

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
extra_words = ['']
stopwords_extra = STOPWORDS.update(extra_words)
```

```python
df_path = '../dataframes/genres/cleaned_combined.csv'
```

```python
df = pd.read_csv(df_path, index_col = 0)
```

```python
df.head()
```

```python
def get_top_15_words_by_frequency(row):
    lyrics = row['Song Lyrics']
    wc = WordCloud(background_color="white", stopwords=STOPWORDS, max_words=15,
              width=800, height=400, collocations=False)
    wc.generate(lyrics)
    words_fc = pd.Series(wc.words_)
    top_15_words = list(words_fc[:15].index)
    top_15_string = ' '.join(top_15_words)
    return top_15_string
```

```python
def apply_top_15_words(df):
    df['top15'] = df.apply(
        lambda row: get_top_15_words_by_frequency(row),
        axis = 1
    )
    return df
```

```python
artists = list(df['Artist'].unique())
```

```python
top_15_df = pd.DataFrame(columns = ['Artist', 'top15words'])
```

```python
for artist in artists:
    artist_df = df[df['Artist'] == artist]
    artist_top15 = ' '.join(df['top15'].str.lower())
    new_row = {
        'Artist': artist,
        'top15words': artist_top15
    }
    new_row_df = pd.DataFrame([new_row])
    new_row_df.index = [artist]

    top_15_df = pd.concat([top_15_df, new_row_df])
```
