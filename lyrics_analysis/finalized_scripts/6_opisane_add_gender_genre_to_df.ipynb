### decades - by looking at names in respective decade dataframes

```python
dfs_path = '../decade_dataframes/decades/'
```

```python
df_80s = pd.read_csv(dfs_path + 'rap_1980s_combined_10.csv')
df_90s = pd.read_csv(dfs_path + 'rap_1990s_combined_9.csv')
df_00s = pd.read_csv(dfs_path + 'rap_2000s_combined_10.csv')
df_10s = pd.read_csv(dfs_path + 'rap_2010s_combined_9.csv')
df_20s = pd.read_csv(dfs_path + 'rap_2020s_combined_10.csv')
```

```python
d_1980s = list(df_80s['Artist'].unique())
d_1990s = list(df_90s['Artist'].unique())
d_2000s = list(df_00s['Artist'].unique())
d_2010s = list(df_10s['Artist'].unique())
d_2020s = list(df_20s['Artist'].unique())
```

```python
decade_lists = [
    d_1980s,
    d_1990s,
    d_2000s,
    d_2010s,
    d_2020s,
]
decade_names = [
    '1980s',
    '1990s',
    '2000s',
    '2010s',
    '2020s',
]
```

```python
def add_decade(df):
    """
    adds decade column to supplied df, returns df with new column
    """
    df['decade'] = '1980s'
    for i in df.index:
        for decade_i, decade_list in enumerate(decade_lists):
            if df['Artist'][i] in decade_list:
                df['decade'][i] = decade_names[decade_i]
    return df
```

```python
### singular df file
df_path = '../decade_dataframes/artist_groupby_sum.csv'
decade_df = pd.read_csv(df_path)
# decade_df = pd.read_csv('../decade_dataframes/decades_counts.csv', index_col=0)
# for this file too
# decade_df = pd.read_csv('../decade_dataframes/cleaned_df.csv', index_col=0)
```

```python
decade_df = add_decade(decade_df)
```

```python
def add_gender(df):
    """
    adds genre column to supllied df, returns df with new column
    """
    df['gender'] = 'male'
    for i in df.index:
        if df['Artist'][i].lower() in male_artists:
            df['gender'][i] = 'male'
        if df['Artist'][i].lower() in female_artists:
            df['gender'][i] = 'female'
    return df
```

### Add extra genre data

```python
rap_artists = [
    'jayz',
    'kendrick lamar',
    'mf doom',
    'kanye west',
    'eminem',
    '2pac',
    'ice cube',
    'nas',
    'a tribe called quest',
    'mos def'
]

rock_artists = [
    'queens of the stone age',
    'nirvana',
    'foo fighters',
    'the strokes',
    'the white stripes',
    'led zeppelin',
    'fugazi',
    'the rolling stones',
    'black sabbath',
    'deep purple'
]

pop_artists = [
    'taylor swift',
    'dua lipa',
    'carly rae jepsen',
    'madonna',
    'michael jackson',
    'charli xcx',
    'olivia rodrigo',
    'beyonc√©',
    'aaliyah',
    'britney spears'
]

soul_artists = [
    'stevie wonder',
    'marvin gaye',
    'curtis mayfield',
    'otis redding',
    'aretha franklin',
    'al green',
    'nina simone',
    'sam cooke',
    'clarence carter',
    'minnie riperton'
]
```

```python
def add_genre(df):
    df['genre'] = 'rap'
    for i in df.index:
        if df['Artist'][i].lower() in rap_artists:
            df['genre'][i] = 'rap'
        if df['Artist'][i].lower() in rock_artists:
            df['genre'][i] = 'rock'
        if df['Artist'][i].lower() in pop_artists:
            df['genre'][i] = 'pop'
        if df['Artist'][i].lower() in soul_artists:
            df['genre'][i] = 'soul'
    return df
```
