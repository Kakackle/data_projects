### Add artist gender data

```python
male_artists = [
    'jayz', 'kendrick lamar', 'mf doom', 'kanye west',
    'eminem', '2pac', 'ice cube', 'nas', 'a tribe called quest', 'mos def',

    'queens of the stone age', 'nirvana', 'foo fighters', 'the strokes',
    'the white stripes', 'led zeppelin', 'fugazi', 'the rolling stones',
    'black sabbath', 'deep purple',

    'michael jackson',

    'stevie wonder', 'marvin gaye', 'curtis mayfield', 'otis redding',
    'al green', 'sam cooke', 'clarence carter'
]

female_artists = [
    'taylor swift', 'dua lipa', 'carly rae jepsen', 'madonna', 'charli xcx',
    'olivia rodrigo', 'beyonc√©', 'aaliyah', 'britney spears',

    'aretha franklin', 'nina simone', 'minnie riperton'
]
```

```python
def add_gender(df):
    """
    adds genre column to supllied df, returns df with new column
    """
    df['gender'] = 'male'
    for i in df.index:
        if df['Artist'][i].lower() in male_artists:
            df['gender'][i] = 'male'
        if df['Artist'][i].lower() in female_artists:
            df['gender'][i] = 'female'
    return df
```

```python
dataframe_path = '../dataframes/genres'
```

```python
for genre in ['pop', 'rock', 'rap', 'soul']:
    df_path = f'{dataframe_path}/cleaned_{genre}_10.csv'

    df = pd.read_csv(df_path, index_col = 0)

    df = add_gender(df)
    df.to_csv(df_path)
```

### Add extra genre data

```python
rap_artists = [
    'jayz',
    'kendrick lamar',
    'mf doom',
    'kanye west',
    'eminem',
    '2pac',
    'ice cube',
    'nas',
    'a tribe called quest',
    'mos def'
]

rock_artists = [
    'queens of the stone age',
    'nirvana',
    'foo fighters',
    'the strokes',
    'the white stripes',
    'led zeppelin',
    'fugazi',
    'the rolling stones',
    'black sabbath',
    'deep purple'
]

pop_artists = [
    'taylor swift',
    'dua lipa',
    'carly rae jepsen',
    'madonna',
    'michael jackson',
    'charli xcx',
    'olivia rodrigo',
    'beyonc√©',
    'aaliyah',
    'britney spears'
]

soul_artists = [
    'stevie wonder',
    'marvin gaye',
    'curtis mayfield',
    'otis redding',
    'aretha franklin',
    'al green',
    'nina simone',
    'sam cooke',
    'clarence carter',
    'minnie riperton'
]
```

```python
def add_genre(df):
    df['genre'] = 'rap'
    for i in df.index:
        if df['Artist'][i].lower() in rap_artists:
            df['genre'][i] = 'rap'
        if df['Artist'][i].lower() in rock_artists:
            df['genre'][i] = 'rock'
        if df['Artist'][i].lower() in pop_artists:
            df['genre'][i] = 'pop'
        if df['Artist'][i].lower() in soul_artists:
            df['genre'][i] = 'soul'
    return df
```

```python
for genre in ['pop', 'rock', 'rap', 'soul']:
    df_path = f'{dataframe_path}/cleaned_{genre}_10.csv'

    df = pd.read_csv(df_path, index_col = 0)

    df = add_gender(df)
    df.to_csv(df_path)
```
