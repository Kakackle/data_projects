```python
import wordcloud
from wordcloud import WordCloud, STOPWORDS

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(df_path, index_col = 0)
```

### Functions to calculate top 15 words for every song and add that data to the dataframe

```python
def get_top_15_words_by_frequency(row):
    lyrics = row['Song Lyrics']
    wc = WordCloud(background_color="white", stopwords=STOPWORDS, max_words=15,
              width=800, height=400, collocations=False)
    wc.generate(lyrics)
    words_fc = pd.Series(wc.words_)
    top_15_words = list(words_fc[:15].index)
    top_15_string = ' '.join(top_15_words)
    return top_15_string
```

```python
def apply_top_15_words(df):
    df['top15'] = df.apply(
        lambda row: get_top_15_words_by_frequency(row),
        axis = 1
    )
    return df
```

```python
artists = list(df['Artist'].unique())
```

#### Add row with top 15 words to df with artist lyrics and data
based on getting top 15 words of top 15 words instead of the whole artist corpus - kinda pointless? keep ti for now for comparison

```python
apply_top_15_words(df)
```

```python
top_15_df = pd.DataFrame(columns = ['Artist', 'genre', 'gender', 'top15words'])
```

```python
for artist in artists:
    artist_df = df[df['Artist'] == artist]
    artist_df.reset_index(inplace=True)
    artist_top15 = ' '.join(artist_df['top15'].str.lower())
    new_row = {
        'Artist': artist,
        'genre': artist_df['genre'][0],
        'gender': artist_df['gender'][0],
        'top15words': artist_top15
    }
    new_row_df = pd.DataFrame([new_row])
    new_row_df.index = [artist]

    top_15_df = pd.concat([top_15_df, new_row_df])
```

```python
top_15_df
```

```python
for index, artist in enumerate(top_15_df.sample(4).iterrows()):
    # print(type(artist))
    # print(artist)
    words = artist[1]['top15words'].lower()
    # print('words')
    # print(words)
    
    sum_wc = WordCloud(background_color="white", stopwords=STOPWORDS, max_words=150,
              width=800, height=400, collocations=False)
    sum_wc.generate(words)

    plt.subplot(2,2, index+1)
    
    plt.imshow(sum_wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'{artist[1]["Artist"]} wc')
```

```python
top_15_df.to_csv(f'../dataframes/top_15_of_top_15_words.csv')
```

### Top 15 by whole artist corpus
a little better

```python
corpus_df = pd.DataFrame(columns = ['Artist', 'genre', 'gender', 'top15words'])
for artist in artists:
    artist_df = df[df['Artist'] == artist]
    lyrics = ' '.join(artist_df['Song Lyrics'].str.lower())
    wc = WordCloud(stopwords = STOPWORDS, max_words=15, collocations=False)
    wc.generate(lyrics)
    words_fc = pd.Series(wc.words_)
    top_15_words = list(words_fc[:15].index)
    top_15_string = ' '.join(top_15_words)

    artist_df.reset_index(inplace=True)
    new_row = {
        'Artist': artist,
        'genre': artist_df['genre'][0],
        'gender': artist_df['gender'][0],
        'top15words': top_15_string
    }
    new_row_df = pd.DataFrame([new_row])
    new_row_df.index = [artist]
    
    corpus_df = pd.concat([corpus_df, new_row_df])
```

```python
for index, artist in enumerate(corpus_df.sample(4).iterrows()):
    # print(type(artist))
    # print(artist)
    words = artist[1]['top15words'].lower()
    # print('words')
    # print(words)
    
    sum_wc = WordCloud(background_color="white", stopwords=STOPWORDS, max_words=150,
              width=800, height=400, collocations=False)
    sum_wc.generate(words)

    plt.subplot(2,2, index+1)
    
    plt.imshow(sum_wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'{artist[1]["Artist"]} wc')
```
